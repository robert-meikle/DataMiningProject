```python
import pandas as pd
import matplotlib.pyplot as plt

data_path = "../data/"
data_files = ["AAPL", "BRK_B", "DVN", "NVDA", "PCG", "ROST", "V", "VNQ"]

vis_path = "../visualizations/"
```

```python
for f in data_files:
	df = pd.read_csv(data_path + f + ".csv", parse_dates=["Date"], index_col="Date")

	df[['Close/Last']] = df[['Close/Last']].replace('[\$,]','',regex=True).astype(float)
	df[['Open']] = df[['Open']].replace('[\$,]','',regex=True).astype(float)
	df[['High']] = df[['High']].replace('[\$,]','',regex=True).astype(float)
	df[['Low']] = df[['Low']].replace('[\$,]','',regex=True).astype(float)
    
	plt.figure(figsize=(10,6), dpi=300)
	plt.plot(df["High"], linewidth=0.25, color='k')
	plt.title(f"{f}, High Price (2013-2023)")
	plt.ylabel("Price (US Dollars)")
	plt.savefig(vis_path + f"{f}.svg", bbox_inches="tight", format='svg')
```
